```python
# check dor duplicates

def check_duplicates(df):
  return df.duplicated().sum()

print("Duplicates: ", check_duplicates(df))

# check missing values

def check_missing_values(df):
  return df.isnull().sum()

print(check_missing_values(df))
```

```python
# Remove rows with missing values in the rating_count column
df.dropna(subset=['rating_count'], inplace=True)
print(check_missing_values(df))
```

```python
# check datatypes
def check_datatypes(df):
  return df.dtypes

print(check_datatypes(df))
```

```python
# Perfrom transformations

# 1. Convert discounted_price to numeric
df['discounted_price'] = df['discounted_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# 2. Convert actual_price to numeric
df['actual_price'] = df['actual_price'].astype(str).str.replace(r'[^\d.]', '', regex=True)
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# 3. Convert discount percentage to numeric
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '', regex=False)
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# 4. Convert rating to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
# check converted datatypes
print(check_datatypes(df))
```

```python
# create new columns that might help in recommendation

# 1. create a new column called weighted_ratings
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df['weighted_ratings'] = df['rating'] * df['rating_count']


# 2. To know more about the product and category, create two more coulmns as main_category and sub_category
df['main_category'] = df['category'].str.split('|').str[0]
df['sub_category'] = df['category'].str.split('|').str[-1]

# print all columns
print(df.columns)
```

# RECOMMENDATION SYSTEM
### Now comes the main part of the project:

recommendation system to help users discover products they will love. To achieve this, I am using a dataset from Amazon, which contains detailed information about products, reviews, and users.

The dataset has 1,462 rows and includes key columns such as 'product_name', 'category', 'discounted_price', 'actual_price', and more. Using this data, I aim to create a personalized recommendation system that suggests products tailored to a user's interests and past purchases.

When a user logs in, they will receive recommendations based on their previous interactions with the platform. This system will enhance the shopping experience by helping users find relevant and exciting products conveniently and efficiently.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['user_id_encoded'] = le.fit_transform(df['user_id'])

# Create a new dataframe with the user_id frequency table
freq_table = pd.DataFrame({'User ID': df['user_id_encoded'].value_counts().index, 'Frequency': df['user_id_encoded'].value_counts().values})

# Display the dataframe
print(freq_table)
id_example = freq_table.iloc[0,0]
print(id_example)
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def recommend_products(df, user_id, top_n=5):
    """
    Recommends products to a user based on their browsing history.

    Args:
        df: The dataframe containing product information and user interactions.
        user_id: The ID of the user for whom to generate recommendations.
        top_n: The number of top recommendations to return.

    Returns:
        A list of recommended product names, or a message if the user ID is not found.
    """

    # 1. Prepare the data
    # Assuming 'product_name' and 'category' are relevant features
    df['combined_features'] = df['product_name'] + ' ' + df['category']

    # 2. Vectorize product features
    tfidf = TfidfVectorizer()
    tfidf_matrix = tfidf.fit_transform(df['combined_features'])

    # 3. Find similar products based on cosine similarity
    cosine_similarities = cosine_similarity(tfidf_matrix)

    # Get the index of the current product
    # Check if the user_id exists in the DataFrame
    if user_id in df['user_id'].values:
        idx = df[df['user_id'] == user_id].index[0]  # Assumes one product per user for now
    else:
        return "Product not found in database for this user_id"  # Return a message if not found


    # Get pairwise similarity scores for the current product
    sim_scores = list(enumerate(cosine_similarities[idx]))

    # Sort the products based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)


    # Get the scores of the 10 most similar products
    sim_scores = sim_scores[1:top_n+1]  # Exclude the product itself (index 0)

    # Get the product indices
    product_indices = [i[0] for i in sim_scores]

    # Return the top N most similar product names
    return df['product_name'].iloc[product_indices].tolist()


# Example usage: Assuming you have a user_id 'user123'
# Replace 'user123' with actual user ID from your dataframe, making sure it exists
recommendations = recommend_products(df, 'AG3D6O4STAQKAY2UVGEUV46KN35Q,AHMY5CWJMMK5BJRBBSNLYT3ONILA,AHCTC6ULH4XB6YHDY6PCH2R772LQ,AGYHHIERNXKA6P5T7CZLXKVPT7IQ,AG4OGOFWXJZTQ2HKYIOCOY3KXF2Q,AENGU523SXMOS7JPDTW52PNNVWGQ,AEQJHCVTNINBS4FKTBGQRQTGTE5Q,AFC3FFC5PKFF5PMA52S3VCHOZ5FQ')  # Replaced 'user123' with a valid user ID
print("Recommended Products:")
for recommendation in recommendations:
    print(recommendation)
```
